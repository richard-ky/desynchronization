#Data Preprocessing

```python
# Mount drive for quick access to files

from google.colab import drive
drive.mount('/content/drive')
```

```python
# Import data

import scipy

temp = scipy.io.loadmat('/content/drive/MyDrive/Data Collection Neural Network Synchronization/Desynchronization/Data/DesyncneuraldataTrial_1(10,0.1,0.1,10,5,450).mat')
```

```python
import pandas as pd
```

```python
# Create new DataFrame

df = pd.DataFrame()
```

```python
# Check shape

print(temp['Input_data'].shape)
print(temp['Label_data'].shape)
print(temp['spike_time_input'].shape)
print(temp['spike_time_label'].shape)
```

```python
# Add data to DataFrame

df['input'] = temp['Input_data'][8:, 0]
df['label'] = temp['Label_data'][1:, 0]
df.head()
```

```python
# Create a row of all the non-lambda features 

new = pd.DataFrame(temp['Input_data'][2:8, 0]).transpose()
new
```

```python
# Create the column names

new.columns = ['Amp', 'PW1', 'PW2', 'PW3', 'Ns', 'Re1']
new
```

```python
# Duplicate the row

new = pd.concat([new]*1600, ignore_index=True)
new.head()
```

```python
# Add non-lambda features to DataFrame

df[['Amp', 'PW1', 'PW2', 'PW3', 'Ns', 'Re1']] = new[['Amp', 'PW1', 'PW2', 'PW3', 'Ns', 'Re1']]
```

```python
# Rearrange columns so 'label' is last

df = df[['input', 'Amp', 'PW1', 'PW2', 'PW3', 'Ns', 'Re1', 'label']]
df.head()
```

```python
# Check counts

print(df['label'].unique())
print(df['label'].value_counts())
```

#Models

```python
# Split data into train and test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['input', 'Amp', 'PW1', 'PW2', 'PW3', 'Ns', 'Re1']], df['label'], test_size=.2)
```

```python
# Support vector classifier

from sklearn.svm import SVC
from sklearn import metrics

svm_classifier = SVC()
svm_classifier.fit(X_train, y_train)

pred = svm_classifier.predict(X_test)
print('Accuracy: ', metrics.accuracy_score(y_test, pred))
```
